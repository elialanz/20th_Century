```python
# Import Libraries

from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

### 3. Load the twentieth-century data that you scraped in Exercise 1.4.

```python
# Import Txt File

with open('key_events_20th_century.txt', 'r', errors='ignore') as file: 
   data = file.read().replace( '\n', ' ')
```

### 4. Tokenize the words from the text and create a bar chart to plot the 10 most common words.

```python
# Download required NLTK tokenization resources

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Sentence Tokenization

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word Tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# Compute and display the frequency distribution of all tokenized words

from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
# Show the 10 most frequent words in the distribution

dist_words.most_common(10)
```

```python
# Frequency Distribution Plot - Plotting the Bar Chart

plt.figure(figsize=(8, 3))
sns.barplot(
    x=[w for w, c in dist_words.most_common(10)],
    y=[c for w, c in dist_words.most_common(10)],
    palette=sns.dark_palette("xkcd:blue", 10)
)
plt.show()
```

### 5. Remove stop words (and punctuation marks if necessary), and rerun the plot. What has changed?
### Comment on your impressions in a markdown cell under the plot.

```python
# Import StopWords

from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
# Removing stopwords in words

filtered_words = [] # creates an empty list
for word in tokenized_word:
      if word not in stop_words:
           filtered_words.append(word)
```

```python
# Create a new FreqDist for filtered–words

dist_words_filter = FreqDist(filtered_words)
print (dist_words_filter)
```

```python
# Frequency Distribution Plot – Plotting The New Cleaned Bar Chart

plt.figure(figsize=(8, 3))
sns.barplot(
    x=[w for w, c in dist_words_filter.most_common(10)],
    y=[c for w, c in dist_words_filter.most_common(10)],
    palette=sns.dark_palette("xkcd:blue", 10)
)
plt.show()
```

#### This is not a clean result yet, i need to exclude punctuation marks aswell.

```python
# Substitute all punctuation marks with a space

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                        " ",        # Replace all non-letters with spaces
                        str(filtered_words))
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print (tokenized_word_2)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(30, cumulative=False)
plt.show()
```

#### I still have some stopwords, i need to filter more

```python
# Build custom stopword filter

new_stopwords = ["The", "s", "th", "In", "II", "Il", "S"]
```

```python
# Filter tokenized words again

filtered = []
for word in tokenized_word_2:
    if word not in new_stopwords:
        filtered.append(word)
```

```python
# Make a new FreqDist

dist_words_filter_3 = FreqDist(filtered)
```

```python
# Replot The New Cleaned Frequency Distribution Bar Chart

plt.figure(figsize=(8, 3))
sns.barplot(
    x=[w for w, c in dist_words_filter_3.most_common(10)],
    y=[c for w, c in dist_words_filter_3.most_common(10)],
    palette=sns.dark_palette("xkcd:blue", 10)
)
plt.show()
```

### 9. Create three bar plots with the top 15 POS labels, one each for nouns, verbs, and adjectives.
### Analyzing the results, what do you think stands out in these frequencies? Comment on your impressions in a Markdown cell under the plot.

```python
# Display the column names of the df_text dataframe

df_text.columns
```

```python
# Clean up the leading single quotes from ALL words

df_text['Words'] = df_text['Words'].str.replace("'", "", regex=False)
```

```python
# Filter nouns, count occurrences, and extract the top 15 most frequent ones

df_nouns = df_text[
    (df_text['Word type'] == "NN") |
    (df_text['Word type'] == "NNS") |
    (df_text['Word type'] == "NNP")
].copy()

df_nouns.columns = ["Word", "Occurrences"]    # same trick as in the lesson

x = df_nouns.groupby('Word').count().reset_index()
y = x.sort_values(by='Occurrences', ascending=False)

top15_nouns = y.nlargest(15, 'Occurrences')
```

```python
# Plot a bar chart showing the most frequent nouns in the dataset

plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(
        x="Word",
        y="Occurrences",
        data=top15_nouns,
        saturation=0.9,
        palette=sns.dark_palette("xkcd:blue", 15)
    ).set_title("20th Century Key Events – most frequently used nouns")

plt.xticks(rotation=60, ha='right')   # <-- ONLY ADDITION
plt.tight_layout()                    # <-- prevents label cutoff
plt.show()
```

#### From here i can see pretty much something similar to the previous charts, War stands out along with Soviet, Hitler, German and Germans, therefore reflecting the true events on world war II that happened.

```python
# repeat the same structure for verbs (VB, VBD, VBG, VBN, VBP, VBZ)

# Verbs only
df_verbs = df_text[
    (df_text['Word type'] == "VBD") |
    (df_text['Word type'] == "VBG")
].copy()

df_verbs.columns = ["Word", "Occurrences"]    # same trick as in the lesson

x = df_verbs.groupby('Word').count().reset_index()
y = x.sort_values(by='Occurrences', ascending=False)

top15_verbs = y.nlargest(15, 'Occurrences')
```

### 10. Create a dataframe with the countries and the number of times they’re mentioned in your text (remember—you’ll need to use the countries lookup list if you didn’t scrape one in the previous task).

```python
# Assigning Path

import os
path = r'/Users/elia/Desktop'
```

```python
# Join the filtered list into a single string and print it

listToStr = ' '.join([str(elem) for elem in filtered])

print(listToStr)
```

```python
# Import the regular expressions module for text pattern processing (Needed for counting two words name countries)

import re
```

```python
# Build a clean, lowercased text string

text_str = " ".join([w.lower() for w in tokenized_word_2])  # or filtered
```

```python
# Load countries and clean names (strip spaces + lowercase)

path = r'/Users/elia/Desktop'

countries = pd.read_csv(os.path.join(path, '20th_Century', 'countries_list.csv'))

country_list = countries['country_name'].str.strip().str.lower().tolist()
```

```python
# Count full phrase occurrences for each country

dict_country_counts = {}
for c in country_list:
    pattern = r'\b' + re.escape(c) + r'\b'   # match whole phrase as words
    count = len(re.findall(pattern, text_str))
    dict_country_counts[c] = count
```

```python
# Build dataframe

df_countries = pd.DataFrame({
    "Country": [c.title() for c in dict_country_counts.keys()],
    "Times mentioned": list(dict_country_counts.values())
})
```

```python
# Sort and show only countries mentioned at least once

df_countries = df_countries.sort_values("Times mentioned",
                                        ascending=False).reset_index(drop=True)

df_countries[df_countries["Times mentioned"] > 0].head(30)
```
