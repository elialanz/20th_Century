```python
import pandas as pd
```

```python
data = pd.DataFrame({
    "Product": ["Laptop", "Headphones", "Coffee Mug", "Backpack"],
    "Category": ["Electronics", "Accessories", "Kitchen", "Travel"],
    "Price": [1299, 199, 15, 89]
})

data
```

```python
# Create a new column by subtracting 10 from each price

data["Discounted"] = data["Price"].apply(lambda x: x - 10)
```

```python
data
```

```python
# Creating a new column for check if products are tech or not

data["Category_Flag"] = data["Category"].apply(
    lambda x: "Tech" if x in ["Electronics", "Accessories"] else "Non-Tech"
)
```
