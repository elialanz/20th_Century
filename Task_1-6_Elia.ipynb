### 6. Split the sentence entities from the NER object.

```python
# Build a dataframe where each row contains a sentence and its extracted named entities

df_sentences = []

# Loop through sentences, get entity list for each sentence
for sent in events_cleaned.sents:
       entity_list = [ent.text for ent in sent.ents]
       df_sentences.append({"sentence": sent, "entities": entity_list})

df_sentences = pd.DataFrame(df_sentences)
```

```python
df_sentences.head(10)
```

```python
# Load The Cleaned Countries Dataframe

countries_cleaned = pd.read_csv("countries_list_clean.csv")
```

```python
# Function to filter out entities not of interest

def filter_entity(ent_list, countries_cleaned):
       return [ent for ent in ent_list
                  if ent in list(countries_cleaned['country_lower'])]
```

```python
# Extract only country-related entities from each sentence’s entity list

df_sentences['countries_entities'] = df_sentences['entities'].apply(lambda x: filter_entity(x, countries_cleaned))
```

### 7. Filter the entities so that you end up only with the ones from your countries list.

```python
# Filter out sentences that don’t have any countries entities

df_sentences_filtered = df_sentences[df_sentences['countries_entities'].map(len) > 0]

df_sentences_filtered.tail(10)
```

### 8. Create the relationships dataframe.

```python
# Defining relationships

window_size = 5
relationships = []

for i in range(df_sentences_filtered.index[-1]):
    end_i = min(i + window_size, df_sentences_filtered.index[-1])
    
    # merge the country lists across the window
    char_list = sum((df_sentences_filtered.loc[i:end_i, "countries_entities"]), [])
    
    # remove duplicates that are NEXT to each other
    char_unique = [char_list[j] for j in range(len(char_list))
                   if (j == 0) or char_list[j] != char_list[j-1]]

    if len(char_unique) > 1:
        for idx, a in enumerate(char_unique[:-1]):
            b = char_unique[idx + 1]
            relationships.append({"source": a, "target": b})
```

```python
# Sort each source–target pair alphabetically to unify reversed duplicates

relationship_df = pd.DataFrame(relationships)

relationship_df
```

```python
# Alphabetically sort each source–target pair to standardize relationship ordering

relationships_df = pd.DataFrame(np.sort(relationship_df.values, axis = 1), columns = relationship_df.columns)
relationships_df.head(5)
```

```python
# Set relationship weights to 1, then group by source–target pairs and sum them

relationships_df["value"] = 1
relationship_df = relationship_df.groupby(["source","target"], sort=False, as_index=False).sum()

relationship_df.head(10)
```

```python
# Add VALUE column

relationships_df["value"] = 1
```

```python
# Group by source–target to count interactions

relationship_df = relationships_df.groupby(["source", "target"], 
                                           sort=False, 
                                           as_index=False).sum()
```
